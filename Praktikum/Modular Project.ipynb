Привести строку `year_of_release` к типу `int` невозможно, поскольку это некоторая особенность библиотеки Pandas, связанная с отсутствием `NA` значения в NumPy и Python изначально. Поэтому, можно перевести это число в тип, к примеру, `Int16`.

```python
df['year_of_release'] = df['year_of_release'].astype('Int16')
df
```

# Как менялись продажи по платформам

```python
df['sum'] = df[['eu_sales', 'jp_sales', 'na_sales',  'other_sales']].sum(axis=1)
df
```

```python
pl_y = df.pivot_table(index=['platform'], values=['sum'], columns='year_of_release', aggfunc=np.sum)
pl_y
```

```python
top_pls = pl_y.sum(axis=1).sort_values(ascending=False).index
```

```python
pl_y = pl_y.reindex(index=top_pls)
```

```python
top_pl_y = pl_y.head(5)['sum']
top_pl_y
```

```python
ax = top_pl_y.transpose().plot.area(alpha=0.5, stacked=False, figsize=(15, 8))
ax.xaxis.set_ticks(range(1980, 2017))
ax.tick_params('x', labelrotation=45)
plt.show()
```

```python
top_pl_y.transpose().describe()
```

# Актуальные данные

```python
year_lims = [2007, 2016]
```

```python
pl_y = df.query('year_of_release >= @year_lims[0] and year_of_release <= @year_lims[1]').pivot_table(index=['platform'], values=['sum'], columns='year_of_release', aggfunc=np.sum)
pl_y
```

```python
top_pls = pl_y.sum(axis=1).sort_values(ascending=False).index
pl_y = pl_y.reindex(index=top_pls)
top_pl_y = pl_y.head(10)['sum']
```

```python
ax = top_pl_y.transpose().plot.area(alpha=0.5, stacked=False, figsize=(15, 8))
ax.xaxis.set_ticks(range(year_lims[0], year_lims[1]+1))
ax.tick_params('x', labelrotation=45)
plt.show()
```

```python
top_platform = ['PS4', 'XOne', 'PS3', 'X360', '3DS']
```

```python
current_df = df.query('year_of_release >= @year_lims[0] and year_of_release <= @year_lims[1]')
current_df
```

```python
top_platform_df = current_df.query('platform in @top_platform')
top_platform_df
```

```python
top_platform_df.boxplot(column=['sum'], by='platform', figsize=(15, 15))
plt.show()
```
