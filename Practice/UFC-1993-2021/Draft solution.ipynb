```python
def get_columns_names(df, columns):
    names_columns = []
    for feature in df.drop('Winner', axis=1).columns:
        if re.search(fr'(_|\b)({"|".join(columns)})(_|\b)', feature, re.IGNORECASE):
            names_columns.append(feature)
            
    return names_columns
```
