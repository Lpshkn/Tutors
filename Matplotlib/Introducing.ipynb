# Описание

**Matplotlib** - библиотека на языке программирования Python для визуализации данных двумерной (2D) графикой (3D графика также поддерживается).

Не в нейросетке счастье - а в том, как правильно обработать данные. Но чтобы их обработатть, необходимо сначала проанализировать корреляции, выбрать нужные данные, выкинуть ненужные и так далее. Для подобных целей часть используется визуализация с помощью библиотеки matplotlib.

Matplotlib может выводить графику, используя различные бэкенд графические библиотеки, как Tk, wxPython. Когда запускается python-скрипт в командной строке, то вся графика обычно отрисовывается в отдельном графическом окне. В Jupyter Notebook имеется возможность отрисовывать графику прямо внутри ноутбука(тетради), используя магическую команду `%matplotlib inline`.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

# Функция plot

Описываемая функция позволяет строить обычные и самые простые графики. На вход она получается список значений, при этом если передан лишь один массив, то она использует эти значения, как значения на вертикальной оси (ось Y), а индексы элементов этого списка, как значения на горизонтальной оси (оси X).

```python
plt.plot([1, 2, 50, 20, 100, 1000, 30])
plt.show()
```

Если ей на вход передаётся два массива значений, то она будет их использовать попарно, как координаты точек на осях X и Y. При этом, первый список - ось X, второй список - ось Y.

```python
plt.plot([0, 1, 5], [100, 50, 205])
plt.show()
```

Оси графика подстраиваются автоматически под передаваемые значения. Однако, их можно самостоятельно конфигурировать, меняя протяженность каждой оси. Это можно сделать, передав в функцию `plt.axis` список `[xmin, xmax, ymin, ymax]`.

```python
plt.plot([0, 1, 5], [100, 50, 205])
plt.axis([-5, 100, 40, 500])
plt.show()
```

На графиках можно строить различные математические функции, которые напрямую зависят от **x**, скажем, функция `y = k*x + b`. Для этого необходимо передать ей список значений x.

Составить подобный список можно с помощью функции `numpy.linspace`, которая вернет k float-переменных, равномерно распределенных в промежутке от a до b.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x ** 2

plt.plot(x, y)
plt.show()
```

Графики можно оформлять. Можно добавить в график оглавление с помощью функции `plt.title`, можно добавить подписи к осям X и Y с помощью функций `plt.xlabel` и `plt.ylabel`. 

Включить сетку можно с помощью функции `plt.grid(True)`.

```python
plt.plot(x, y)
plt.title('Square function')
plt.xlabel('x')
plt.ylabel('y = x ** 2')
plt.grid(True)
plt.show()
```
