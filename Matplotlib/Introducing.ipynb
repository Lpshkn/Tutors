# Описание

**Matplotlib** - библиотека на языке программирования Python для визуализации данных двумерной (2D) графикой (3D графика также поддерживается).

Не в нейросетке счастье - а в том, как правильно обработать данные. Но чтобы их обработатть, необходимо сначала проанализировать корреляции, выбрать нужные данные, выкинуть ненужные и так далее. Для подобных целей часть используется визуализация с помощью библиотеки matplotlib.

Matplotlib может выводить графику, используя различные бэкенд графические библиотеки, как Tk, wxPython. Когда запускается python-скрипт в командной строке, то вся графика обычно отрисовывается в отдельном графическом окне. В Jupyter Notebook имеется возможность отрисовывать графику прямо внутри ноутбука(тетради), используя магическую команду `%matplotlib inline`.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

# Функция plot

Описываемая функция позволяет строить обычные и самые простые графики. На вход она получается список значений, при этом если передан лишь один массив, то она использует эти значения, как значения на вертикальной оси (ось Y), а индексы элементов этого списка, как значения на горизонтальной оси (оси X).

```python
plt.plot([1, 2, 50, 20, 100, 1000, 30])
plt.show()
```

Если ей на вход передаётся два массива значений, то она будет их использовать попарно, как координаты точек на осях X и Y. При этом, первый список - ось X, второй список - ось Y.

```python
plt.plot([0, 1, 5], [100, 50, 205])
plt.show()
```

**Важно отметить**, что по умолчанию matplotlib рисует линию между последовательностями точек, поэтому в аргументах функций передаются списки **координат**. Между этими точка с соответствующими координатами matplotlib строит линии автоматически. Эти линии также можно конфигурировать, об этом позднее.

## Настройка осей

Оси графика подстраиваются автоматически под передаваемые значения. Однако, их можно самостоятельно конфигурировать, меняя протяженность каждой оси. Это можно сделать, передав в функцию `plt.axis` список `[xmin, xmax, ymin, ymax]`.

```python
plt.plot([0, 1, 5], [100, 50, 205])
plt.axis([-5, 100, 40, 500])
plt.show()
```

## Создание последовательности x'ов для функций

На графиках можно строить различные математические функции, которые напрямую зависят от **x**, скажем, функция `y = k*x + b`. Для этого необходимо передать ей список значений x.

Составить подобный список можно с помощью функции `numpy.linspace`, которая вернет k float-переменных, равномерно распределенных в промежутке от a до b.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x ** 2

plt.plot(x, y)
plt.show()
```

## Небольшое оформление графиков

Графики можно оформлять. Можно добавить в график оглавление с помощью функции `plt.title`, можно добавить подписи к осям X и Y с помощью функций `plt.xlabel` и `plt.ylabel`. 

Включить сетку можно с помощью функции `plt.grid(True)`.

```python
plt.plot(x, y)
plt.title('Square function')
plt.xlabel('x')
plt.ylabel('y = x ** 2')
plt.grid(True)
plt.show()
```

# Стили линий и цвета

Как было сказано ранее, matplotlib автоматически рисует линии между последовательными точками на графике.

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0])
plt.axis([-10, 110, -10, 140])
plt.show()
```

Чтобы изменить **стиль линии** и **цвет**, необходимо передать **третий аргумент** в функцию отрисовки. Например, `g--` означает *зеленая прерывистая линия*.

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

На одном полотне можно отрисовать сразу несколько графиков. Для этого достаточно передать их последовательно в функцию `plot` в формате `x1, y1, [style1], x2, y2, [style2], ...`.

Также, это будет работать в случае, если вызвать несколько функций `plot` отдельно.

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-")
plt.plot([0, 50, 100, 0], [100, 130, 100, 100], '^k-')
plt.plot([0, 100, 0, 100], [0, 100, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

## Основные маркеры, линии и цвета

### Маркеры

|Символ|Описание|
|:----:|:------:|
|   .  |график точками|
|   ,  |график пикселями|
|   o  |кружочками|
|   v  |треугольники вниз|
|   ^  |треугольники вверх|
|   s  |квадратики|
|   *  |звёздочки|
|   +  |плюсики|
|   x  |иксы|

### Линии

|Символ|Описание|
|:----:|:------:|
|   -  |непрерывная линия|
|   -- |прерывистая линия|
|   -. |точка-тире|
|   :  |линия точками|

### Цвета

|Символ|Описание|
|:----:|:------:|
|   b  |синий (blue)|
|   g  |зеленый (green)|
|   r  |красный (red)|
|   c  |голубой (cyan)|
|   m  |пурпурный (magenta)|
|   y  |жёлтый (yellow)|
|   k  |черный (black)|
|   w  |белый (white)|

Важно отметить, что формат можно совмещать, к примеру, `^k:` будет означать черные треугольники вверх, соединённые между собой точечными линиями.

Также, цвет можно передать с помощью строки в шестнадцатеричном ввиде `'#008000'`

```python
x = np.linspace(-1.5, 1.5, 30)
plt.plot(x, x, '^k--', x, x**2, 'vr:', x, x**3, 'bo')
plt.axis([-2, 2, -5, 5])
plt.show()
```

## Конфигурирование линий отдельно

`plot` функция возвращает список `Line2D` объектов (по одному на линию). Можно установить дополнительные атрибуты каждой линии, как например, ширину линии (`set_linewidth`), прозрачность (`set_alpha`) и так далее.

Полное описание можно посмотреть [тут](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html?highlight=line2d#matplotlib.lines.Line2D).

```python
x = np.linspace(-1.5, 1.5, 30)
l1, l2, l3 = plt.plot(x, x, 'k--', x, x**2, 'vr:', x, x**3, 'bo')
l1.set_linewidth(3)
l1.set_dash_capstyle('butt')
l3.set_alpha(0.2)
plt.show()
```
