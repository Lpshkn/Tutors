# Описание

**Matplotlib** - библиотека на языке программирования Python для визуализации данных двумерной (2D) графикой (3D графика также поддерживается).

Не в нейросетке счастье - а в том, как правильно обработать данные. Но чтобы их обработатть, необходимо сначала проанализировать корреляции, выбрать нужные данные, выкинуть ненужные и так далее. Для подобных целей часть используется визуализация с помощью библиотеки matplotlib.

Matplotlib может выводить графику, используя различные бэкенд графические библиотеки, как Tk, wxPython. Когда запускается python-скрипт в командной строке, то вся графика обычно отрисовывается в отдельном графическом окне. В Jupyter Notebook имеется возможность отрисовывать графику прямо внутри ноутбука(тетради), используя магическую команду `%matplotlib inline`.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

# Функция plot

Описываемая функция позволяет строить обычные и самые простые графики. На вход она получается список значений, при этом если передан лишь один массив, то она использует эти значения, как значения на вертикальной оси (ось Y), а индексы элементов этого списка, как значения на горизонтальной оси (оси X).

```python
plt.plot([1, 2, 50, 20, 100, 1000, 30])
plt.show()
```

Если ей на вход передаётся два массива значений, то она будет их использовать попарно, как координаты точек на осях X и Y. При этом, первый список - ось X, второй список - ось Y.

```python
plt.plot([0, 1, 5], [100, 50, 205])
plt.show()
```

**Важно отметить**, что по умолчанию matplotlib рисует линию между последовательностями точек, поэтому в аргументах функций передаются списки **координат**. Между этими точка с соответствующими координатами matplotlib строит линии автоматически. Эти линии также можно конфигурировать, об этом позднее.

## Настройка осей

Оси графика подстраиваются автоматически под передаваемые значения. Однако, их можно самостоятельно конфигурировать, меняя протяженность каждой оси. Это можно сделать, передав в функцию `plt.axis` список `[xmin, xmax, ymin, ymax]`.

```python
plt.plot([0, 1, 5], [100, 50, 205])
plt.axis([-5, 100, 40, 500])
plt.show()
```

## Создание последовательности x'ов для функций

На графиках можно строить различные математические функции, которые напрямую зависят от **x**, скажем, функция `y = k*x + b`. Для этого необходимо передать ей список значений x.

Составить подобный список можно с помощью функции `numpy.linspace`, которая вернет k float-переменных, равномерно распределенных в промежутке от a до b.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x ** 2

plt.plot(x, y)
plt.show()
```

## Небольшое оформление графиков

Графики можно оформлять. Можно добавить в график оглавление с помощью функции `plt.title`, можно добавить подписи к осям X и Y с помощью функций `plt.xlabel` и `plt.ylabel`. 

Включить сетку можно с помощью функции `plt.grid(True)`.

```python
plt.plot(x, y)
plt.title('Square function', fontsize=15)
plt.xlabel('x', fontsize=15)
plt.ylabel('y = x ** 2', fontsize=15)
plt.grid(True)
plt.show()
```

# Стили линий и цвета

Figure  
Axes

Как было сказано ранее, matplotlib автоматически рисует линии между последовательными точками на графике.

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0])
plt.axis([-10, 110, -10, 140])
plt.show()
```

Чтобы изменить **стиль линии** и **цвет**, необходимо передать **третий аргумент** в функцию отрисовки. Например, `g--` означает *зеленая прерывистая линия*.

```python
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

На одном полотне можно отрисовать сразу несколько графиков. Для этого достаточно передать их последовательно в функцию `plot` в формате `x1, y1, [style1], x2, y2, [style2], ...`.

Также, это будет работать в случае, если вызвать несколько функций `plot` отдельно.

```python
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-", [0, 50, 100, 0], [100, 130, 100, 100], '^k-')
#plt.plot([0, 50, 100, 0], [100, 130, 100, 100], '^k-')
plt.plot([0, 100, 0, 100], [0, 100, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

## Конфигурирование линий отдельно

`plot` функция возвращает список `Line2D` объектов (по одному на линию). Можно установить дополнительные атрибуты каждой линии, как например, ширину линии (`set_linewidth`), прозрачность (`set_alpha`) и так далее.

Полное описание можно посмотреть [тут](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html?highlight=line2d#matplotlib.lines.Line2D).

```python
x = np.linspace(-1.5, 1.5, 30)
l1, l2, l3 = plt.plot(x, x, 'k--', x, x**2, 'vr:', x, x**3, 'bo')
l1.set_linewidth(3)
l1.set_dash_capstyle('butt')
l3.set_alpha(0.2)
plt.show()
```

# Сохранение фигуры

Сохранение фигуры в изображение осуществляется с помощью функции `plt.savefig`. Описание [тут](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html?highlight=savefig#matplotlib.pyplot.savefig).

Доступный формат вывода зависист от того, какой backend вы использовали.

**Параметры:**
* **fname** - имя файла, строка или бинарный файл
* **dpi** - float или `figure`, при этом, если число - то разрешение в точках на дюйм, если фигура - то разрешение фигуры. По умолчанию, соотношение сторон области `Figure` составляет 4:3 или 8х6 пространственных единиц, одна такая единица соответствует 100 пикселям и тогда сохраняемая картинка будет иметь размеры 800х600px. Если указать `dpi` = 200, то одной единице будет соответствовать уже 200px, а изображение будет иметь размер 1600х1200px.
* **quality** - применяется только для формата *jpg* и *jpeg*, принимает от 1 до 95. DEPRECATED.
* **transparent** - делает прозрачными области Axes. Полезно, если необходимо разместить картинку поверх презентации.


```python
x = np.linspace(-1.5, 1.5, 30)
plt.plot(x, x**2)
plt.savefig('fig.png', dpi=100, transparent=True)
```

# Subplots или несколько графиков на одной фигуре

Фигура Matplotlib может содержать несколько графиков. Эти subplots организованы при помощи сетки (grid). Чтоб создать дополнительный график, необходимо вызвать функцию `subplot` и указать количество строк и колонок в этой сетке и индекс того графика, на котором хочется рисовать (индексация производится по порядку каждого ряда - слева направа на каждом ряду). 

Отметим, что **pyplot** считает активным выбранный график (который можно получить с помощью `plt.gca()`). Когда мы вызываем функцию `plot`, она рисует в активном окне.

При этом указание колонок и индексы можно указать цельным числом, например `212`.

```python
x = np.linspace(-1.5, 1.5, 30)
plt.figure(figsize=(15, 10))
plt.subplot(2, 1, 1)
plt.plot(x, x)
plt.subplot(212)
plt.plot(x, x**3)
plt.show()
```

Графики можно даже растягивать на несколько столбцов:

```python
plt.subplot(2, 2, 1)
plt.plot(x, x)
plt.subplot(2, 2, 2)
plt.plot(x, x**2)
plt.subplot(2, 1, 2)
plt.plot(x, x**3)
plt.show()
```

Есть и более крутая функция "для прошаренных": `plt.subplot2grid`, которая позволяет задать размер сетки и указать для какой конкретно ячейки (к примеру `(0, 0)`) применять следующую отрисовку. При это можно указать сколько строк и столбцов это займет.

```python
plt.subplot2grid((2, 2), (0, 0), rowspan=1, colspan=1)
plt.plot(x, x)
plt.subplot2grid((2, 2), (0, 1))
plt.plot(x, x**2)
plt.subplot2grid((2, 2), (1, 0), colspan=2)
plt.plot(x, x**3)
plt.show()
```

# Несколько фигур

Также, предоставляется возможность отрисовки **нескольких** фигур. Каждая фигура может содержать один или несколько графиков. По умолчанию, Matplotlib автоматически использует `plt.figure(1)`.

Вы можете захотеть сменить фигуру, тогда pyplot переключит активную фигуру на неё (активную фигуру можно получить функцией `plt.gcf()`.

```python
x = np.linspace(-1.5, 1.5, 30)

plt.figure(1,figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.plot(x, x)
plt.subplot(2, 2, 2)
plt.plot(x, -x**3, 'r:')
plt.subplot(2, 1, 2)
plt.plot(x, x**4, '^k-.')

plt.figure(2)
plt.subplot2grid((2, 2), (0, 0), colspan=2)
plt.plot(x, -x+3, 'vb-')
plt.subplot2grid((2, 2), (1, 0))
plt.plot(x, 1/x, "g--")
plt.grid()
plt.show()
```

# Явное / неявное определение объеков

Каждый раз, когда мы вызываем функцию `plot`, pyplot не просто рисует его на активном полотне, но ещё и создаёт фигуру и само полотно, если они не существуют. Эта магия очень удобна в Jupyter Notebook.

Однако, при написании программ надо избегать неявного определения объектов. 

```python
x = np.linspace(-2, 2, 200)
fig1, (ax_top, ax_bottom) = plt.subplots(2, 1, sharex=True)
fig1.set_size_inches(15, 10)
line1, line2 = ax_top.plot(x, np.sin(x**2), 'r-', x, np.cos(-x**3), 'k:')
line1.set_alpha(0.5)
ax_top.grid(True)
line3 = ax_bottom.plot(x, np.tan(3*x), 'g-')
ax_bottom.grid(False)
ax_bottom.tick_params(labelsize=15)
plt.show()
```

# Отрисовка текста

Вы можете добавить любой текст с помощью функции `plt.text`, передав ей горизонтальную и вертикальную координаты, а также указав необходимые опции. При этом, текст может содержать вставки **TeX**.

```python
x = np.linspace(-1.5, 1.5, 20)
px = 0.5
py = px ** 2
plt.plot(x, x**2, 'y--')
plt.plot(px, py, 'ro')
plt.text(0, 2, 'Функция $y = x^2$', ha='center', fontsize=20, color='green')
plt.text(px-0.5, py, 'Точка', ha='left', weight='bold')
plt.show()
```

Помимо того, чтобы просто отрисовать текст, вы можете ещё и аннотировать какой-либо объект на графике с помощью функции `plt.annotate`. Вы передаёте ей строку, координаты точки и текста + дополнительные опции.

```python
plt.plot(x, x**2, px, py, 'ro')
plt.annotate('Точка', xy=(px, py), xytext=(px-1,py+1), fontsize=14,
             color='green', arrowprops={'facecolor': 'lightgreen'})
plt.show()
```

# Легенды

На график очень легко нанести **легенды**: для этого достаточно определить **label** для каждой линии в функции **plt.plot**. 

```python
plt.plot(x, x**2, 'r:', label="Квадратичная функция")
plt.plot(x, x**3, 'g-.', label="Кубическая функция")
plt.grid(True)
plt.legend(fontsize=14, loc='upper center')
plt.show()
```

# Тики (чёрточки на осях)

Оси имеют маленькие черточки, называемые **тиками**. Подписи рядом с тиками называются **ticklabels**.

```python
plt.figure(1, figsize=(15, 10))
plt.subplot(131)
plt.plot(x, x**3)
plt.grid(True)

ax = plt.subplot(132)
plt.plot(x, x**3)
ax.xaxis.set_ticks([-2, -1, 0, 1, 2])
plt.grid(True)

ax = plt.subplot(133)
ax.xaxis.set_ticks([-2, 0, 1, 2])
ax.yaxis.set_ticks([-3, -2, -1, 0, 1, 2, 3])
ax.yaxis.set_ticklabels(["min", -2, -1, 0, 1, 2, "max"])
plt.plot(x, x**3)
plt.grid(True)

plt.show()
```

# 3D проекция

Создавать 3D проекции очень просто. Для этого необходимо использовать `Axes3D` и в `plt.subplot` указать `projection='3d'`. Функция `subplot` вернет объект `Axes3DSubplot`, который можно использовать для вызова функции plot_surface, передавай ей координаты `x`, `y` и `z`.

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
```

```python
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
```

```python
plt.figure(figsize=(15, 15))
sub = plt.subplot(111, projection='3d')
sub.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=mpl.cm.coolwarm)
plt.show()
```
